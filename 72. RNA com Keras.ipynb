```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, recall_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv('Churn_treino.csv', sep=';')
df.head()
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
standarscaler = StandardScaler()
numerical = X.select_dtypes(include=['int64', 'float64']).columns
# Use the instance of StandardScaler (standardscaler) to call fit_transform
X[numerical] = standarscaler.fit_transform(X[numerical])
```

```python
X
```

```python
labelencoder = LabelEncoder()
categorical = X.select_dtypes(include='object').columns
# Use the instance of LabelEncoder (labelencoder) to call fit_transform
for col in categorical:
    X[col] = labelencoder.fit_transform(X[col])
```

```python
X
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
modelo = Sequential()
modelo.add(Dense(units=64, activation='relu', input_dim=X_treino.shape[1]))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=32, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=32, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit(X_treino, y_treino, epochs=50, batch_size=32)
```

```python
previsoes = modelo.predict(X_teste)
previsoes
```

```python
y_pred = (previsoes > 0.5).astype('int32')
y_pred
```

```python
print('Acurácia: ', accuracy_score(y_teste, y_pred))
print('F1 Score: ', f1_score(y_teste, y_pred))
print('Recall: ', recall_score(y_teste, y_pred))
print('Matriz de confusão: \n', confusion_matrix(y_teste, y_pred))
```
