Converte a classificação multi label em várias classificações binárias ou em classificação multi classe

Classifier Chain

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.multioutput import ClassifierChain
from sklearn.preprocessing import StandardScaler
```

Quando é multilabel cada classe fica em uma coluna

```python
musica = pd.read_csv('Musica.csv')
musica.head(10)
```

```python
classe = musica.iloc[:,0:6].values          #Escolhemos as colunas que contém as classes, ex: amazed-surprised
previsores = musica.iloc[:,7:78].values     #Deixamos as demais colunas dos previsores, como tem 77 colunas, deixamos 78 como final
```

A padronização serve para que todas as variáveis possam contribuir com o modelo

```python
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```

```python
previsores
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=42)     #divisão de dados de treino e teste
```

```python
svm = SVC()
classifier_chain = ClassifierChain(base_estimator=svm, random_state=42)
classifier_chain.fit(X_treinamento, y_treinamento)
```

```python
y_pred = classifier_chain.predict(X_teste)
loss = hamming_loss(y_teste, y_pred)          #Comparo o que aconteceu com o que foi previsto

print(f'Hamming Loss: {loss}')
```

O resultado diz que 19% das previsões foram classificadas erroneamente
