```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
data = h2o.import_file('insurance.csv')
```

```python
data
```

```python
data = data.drop("C1")
```

```python
treino, teste = data.split_frame(ratios=[0.8], seed = 0)
```

```python
model_automl = H2OAutoML(max_runtime_secs=180, sort_metric='AUC')
model_automl.train(y="Accident", training_frame=treino)
```

```python
ranking = model_automl.leaderboard
ranking = ranking.as_data_frame()
ranking
```

Não gerou a AUC pois ela só serve para classificações binárias

```python
prever = model_automl.leader.predict(teste)
prever
```
